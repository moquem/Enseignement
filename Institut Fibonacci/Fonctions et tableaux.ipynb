On peut en Python raliser de manire beaucoup plus efficace  l'aide de __fonctions__. Les fonctions en informatique fonctionne de manire analogue que les fonctions mathmatiques. 
Considrons une fonction mathmatiques : $f(x) = x^2, \forall x \in \mathbb{R}$. On peut remplacer le $x$ par n'importe quelle valeur de $\mathbb{R}$ et on obtient ainsi : $f(3) = 9, f(10)=100$ par exemple. 
En informatique on va de mme dfinir une fonction qui prend en entre un ou plusieurs arguments (le $x$ de $f$) et qui va lui associer une expression ($x^2$). On peut mettre dans l'expression de la fonction n'importe quel lment de syntaxe du langage (des boucles `for`, des tableaux, affectations de variables, etc...).

Dclaration d'une fonction en Python :

```python
def nom_fonction(x1,x2): #nombre d'argument qu'on veut
    """algorithme pouvant utiliser x1, x2, ..."""

#On peut ensuite appeler la fonction sur des valeurs spcifiques.

nom_fonction(3,5)
```

Exemple:

```python
def carre(x):
    print(x**2)

carre(3)
carre(10)
```
