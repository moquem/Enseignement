Dans un programme, on a parfois besoin de réutiliser plusieurs fois un même bout de code. On aimerait alors avoir une manière plus efficace que de recopier tout le morceau de code correspondant pour réaliser notre programme.

__Exemple__ : Ecrire une ligne de 7 ' * ', puis une ligne de 15 ' * ', puis une ligne de 3 ' * '.

```python
for i in range(??):
    print("*",end="") #Permet de ne pas revenir à la ligne après l'étoile.
print() #Pour revenir à la ligne

for i in range(??):
    print("*",end="")
print()

for i in range(??):
    print("*",end="")
print()
```

On constate qu'il faut à chaque fois indiquer le bon nombre dans la boucle `for`, mais qu'on répète exactement le même morceau de code. On peut en Python réaliser de manière beaucoup plus efficace à l'aide de __fonctions__. Les fonctions en informatique fonctionne de manière analogue que les fonctions mathématiques. 
Considérons une fonction mathématiques : $f(x) = x^2, \forall x \in \mathbb{R}$. On peut remplacer le $x$ par n'importe quelle valeur de $\mathbb{R}$ et on obtient ainsi : $f(3) = 9, f(10)=100$ par exemple. 
En informatique on va de même définir une fonction qui prend en entrée un ou plusieurs arguments (le $x$ de $f$) et qui va lui associer une expression ($x^2$). On peut mettre dans l'expression de la fonction n'importe quel élèment de syntaxe du langage (des boucles `for`, des tableaux, affectations de variables, etc...).

Déclaration d'une fonction en Python :

```python
def nom_fonction(x1,x2): #nombre d'argument qu'on veut
    """algorithme pouvant utiliser x1, x2, ..."""

#On peut ensuite appeler la fonction sur des valeurs spécifiques.

nom_fonction(3,5)
```

Exemple:

```python
def carre(x):
    print(x**2)

carre(3)
carre(10)
```

Dans tous les exemples précédents, les fonctions "affichent" quelque chose sur l'ordinateur. Mais on ne peut pas récupérer la valeur des variables calculées à l'intérieur de cette fonction. La majorité du temps, les fonctions sont utiles pour calculer des variables que l'on aimerait réutiliser dans la suite de notre programme. Pour cela on utilise la syntaxe `return`.

```python
def carre(x):
    y = x**2
    return y

#Cette fonction va calculer puis renvoyer le carré d'un réel. On peut stocker le résultat de cette fonction dans une variable dans la suite du programme.

a = carre(3)
b = carre(10)

print(a+b)
```
